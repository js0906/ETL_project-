```python
# dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# sqlite path
database_path = "wine_db.sqlite"
```

```python
# engine connect
engine = create_engine(f"sqlite:///{database_path}")
conn = engine.connect()
```

```python
# read database wine
data_wine = pd.read_sql("SELECT * FROM wine", conn)
data_wine
```

```python
# drop for duplicates
wine_dups = data_wine.drop_duplicates()
wine_dups
```

```python
# check for column names
wine_dups.columns
```

```python
# rename wine

wine_rename = wine_dups.rename(columns={'id': 'ID', 'country': 'Country', 'description': 'Description', 'designation': 'Designation', 'points': 'Rating', 'price': 'Price', 'province': 'Province', 'region_1': 'Primary_Region', 'region_2': 'Secondary_Region', 'variety': 'Variety', 'winery': 'Winery'})
wine_rename
```

```python
# check for column renames
wine_rename.columns
```

```python
# added a $

wine_rename["Price"] = wine_rename["Price"].map("${:,.2f}".format)
wine_rename
```

```python
# drop na for wine
clean_wine = wine_rename.dropna()
clean_wine
```

```python
# read database import

data_imports = pd.read_sql("SELECT * FROM import", conn)
data_imports
```

```python
# drop duplicates of imports
drop_imports = data_imports.drop_duplicates()
drop_imports
```

```python
# drop na for imports
clean_imports = drop_imports.dropna()
clean_imports
```

```python
# export to csv for postgres wine
clean_wine.to_csv("cleaned_wine.csv")
```

```python
# export to csv for postgres
clean_imports.to_csv("cleaned_imports.csv")
```
