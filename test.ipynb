```python
# dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# sqlite path
database_path = "wine_db.sqlite"
```

```python
# engine connect
engine = create_engine(f"sqlite:///{database_path}")
conn = engine.connect()
```

```python
# read database wine
data = pd.read_sql("SELECT * FROM wine", conn)
data
```

```python
# read database import

data_imports = pd.read_sql("SELECT * FROM import", conn)
data_imports
```

```python
# drop duplicates of imports
data_imports.drop_duplicates()
```
